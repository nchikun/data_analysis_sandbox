```python
# 全体のプロット
sns.jointplot(df_cars["Horsepower"], df_cars["Miles_per_Gallon"])
```

```python
# Originごとのプロット
g = sns.FacetGrid(df_cars, hue='Origin', size=5)
g.map(plt.scatter, 'Horsepower', 'Miles_per_Gallon').add_legend()
```

```python
# 全体のプロット
g = sns.pairplot(df_cars, size=2.5)
```

```python
# Originごとのプロット
g = sns.pairplot(df_cars, hue='Origin', size=2.5)
```

```python
# Originごとのカーネル密度推定まで実行
g = sns.pairplot(df_cars, hue='Origin', size=2.5, diag_kind="kde")
```

```python
# 全体の線形回帰まで実行
g = sns.pairplot(df_cars, size=2.5, kind="reg")
```

```python
# Originごとの線形回帰まで実行
g = sns.pairplot(df_cars, hue='Origin', size=2.5, kind="reg")
```
