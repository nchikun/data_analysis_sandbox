```python
import warnings
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
df_1 = pd.DataFrame({
    'age': [24,42,18,68,24,30],
    'state': ['NY','CA','CA','TX','CA','NY'],
    'point': [64,92,70,70,88,57]
})
df_2 = pd.DataFrame({
    'age': [24,42,18,68,24,30],
    'state': ['NY','CA','CA','TX','CA','NY'],
    'point': [64,92,70,70,88,57]
})

# クロス集計表（指定行列による数量表）の作成
ct_1 = pd.crosstab([df_1['age'],df_1['state']], df_1['point'])
ct_2 = pd.crosstab([df_2['age'],df_2['state']], df_2['point'])
display(ct_1)
display(ct_2)

# テーブル同士の演算可能
ct_3 = ct_1/ct_2
ct_3.fillna(0, inplace=True)
display(ct_3)
```

```python
plt.figure(figsize=(15, 10))
sns.set_context('paper', 2.5)
sns.heatmap(ct_1, annot=True, cmap='Blues')
```

```python
plt.figure(figsize=(15, 10))
sns.set_context('paper', 2.5)
sns.heatmap(ct_3, annot=True, cmap='Blues')
```

```python
# 指定行列の含まれる値リストの集計値（数量,平均,標準偏差等）の情報を出力
pt = pd.pivot_table(df_1, index=['age','state'], columns='point', values='age', aggfunc=len)
pt.fillna(0, inplace=True)
display(pt)
plt.figure(figsize=(15, 10))
sns.set_context('paper', 2.5)
sns.heatmap(pt, annot=True, cmap='Blues')
```
